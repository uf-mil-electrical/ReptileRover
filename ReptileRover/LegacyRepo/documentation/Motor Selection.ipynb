# Justification for selection 
Using the following theroms: <br>
$$
P = \dfrac{W}{T}
$$
<br>
$$
W = F*D
$$
We will be applying them to the circular end of the motor



The documentation from source two provides the following information about the motor: <br>
**No-Load Speed: 50RPM**
**Rated Torque: 3.97Kg.cm**
<br>
This infromation lets us re-arrange the prior theroms as thus to calculate power: <br>
$$
P = \dfrac{F*D}{T}
$$
We then utilize our knowns as thus:
$$
\dfrac{D}{TR} = 2*pi*\dfrac{RPM}{60}
$$
<br>
$$
3.97*0.0980665 = .389324005 \dfrac{N}{M} \text{rated torque}
$$


```python
import math
Rated_torque = 3.97
No_load_speed = 50


D_T = 2*math.pi*(No_load_speed/60)
Rated_torque = Rated_torque*0.0980665
Power = Rated_torque*D_T
print(Power)
```

Irregardless of the power what really matters is force generation:
$$
T=F*D
$$
